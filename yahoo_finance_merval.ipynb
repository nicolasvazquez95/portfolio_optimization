## Get info for many stocks


```python
stocks = ['SBUX','AAPL','AXP','KO']

trailing_pe = []
portfolio = []
for stk in stocks:
    try:
        pe = yf.Ticker(stk).get_info()['trailingPE']
        trailing_pe.append((stk, pe))
    except:
        pass

df = pd.DataFrame(trailing_pe, columns=["Stock","PE"])
```

```python
plt.figure(figsize=(20,8))
plt.scatter(df["Stock"],df["PE"], s=df["PE"], c=df["PE"])
plt.hlines(df["PE"].mean(),df.index[0],df.index[-1], linestyle=":")
plt.xticks(rotation=45)
plt.colorbar()
plt.title("Price-to-Earnings Ratio");
```

```python
ko_valuation = yf_si.get_stats_valuation(stock)
```

```python
# Stats for many stocks
stats_df = pd.DataFrame()
stats_df["Ratios"] = ko_valuation["Ratios"]

for stk in stocks:
    stats = yf_si.get_stats_valuation(stk)
    column = stats[stats.columns[1]]
    stats_df[stk] = column
```
